```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.opera import OperaDriverManager
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
import json
import re
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import random
import csv
```

```python

def extract_urls(page_num):
    url = f'https://1000kitap.com/kesfet/en-cok-alinti-yapan-siki-okurlar?sayfa={page_num}'
    time.sleep(2)
        
    # Headless modda çalışacak şekilde Chrome ayarlıyoruz
    options = Options()
    options.headless = True
    driver = webdriver.Chrome(options=options)

    driver.get(url)
    html = driver.page_source
    driver.quit()

    # HTML'i BeautifulSoup ile parse ediyoruz
    soup = BeautifulSoup(html, 'html.parser')

    # Regex ile "ListItem" içeren tüm URL'leri çekiyoruz
    urls = re.findall(r'"url":"(https://1000kitap.com/[^"]+)"', str(soup))

    # URL'leri set'e çevirerek tekrarları engelliyoruz
    unique_urls = set(urls)
    return unique_urls
all_urls = []
for page in range(11, 15):
    urls = extract_urls(page)
    all_urls.extend(urls)
    
```

```python
print(len(all_urls))
print(all_urls)
```

```python
last_link=[]
for user in all_urls:
# Kullanıcı adını almak için "/1000kitap.com/" kısmından sonrasını alıyoruz
    user_name = user.split('/')[-1]
# Yeni linki oluşturuyoruz
    new_link = f'https://1000kitap.com/{user_name}/kitaplari/puanlamalari'
    last_link.append(new_link)

# Sonuçları yazdırıyoruz
for link in last_link:
    print(link)
```

```python
updated_links = [url + "?sayfa=" for url in last_link]
print(updated_links)
```

```python
print(len(updated_links))
#updated_links = updated_links[]
for i in updated_links:
    print(i)
```

```python
csv_file_name = "kitap_verileri.csv"
```

```python
for base_url in updated_links:
    
    options = Options()
    options.headless = True
    driver = webdriver.Chrome(options=options)

    driver.get(base_url)
    html = driver.page_source

    # Kullanıcı sayfasını BeautifulSoup ile kontrol et
    soup = BeautifulSoup(html, 'html.parser')

    
    if not ("Sadece okurun kendisi görebilir" in soup.text or "Sadece karşılıklı takipleştikleri görebilir" in soup.text or"Bu Hesap Gizli" in soup.text or "Sadece kayıtlı okurlar görebilir" in soup.text):
        time.sleep(2)
        current_page = 1
        books_data = []
        all_books_set = set()  # Tüm eklenen kitapları tutan set
        previous_books = set()  # Bir önceki sayfadaki kitapların seti

        while True:
            url = base_url + str(current_page)
            print(f"Sayfa {current_page}: {url}")
            driver.get(url)
            html = driver.page_source
            time.sleep(1)
            # HTML'i BeautifulSoup ile parse ediyoruz
            soup = BeautifulSoup(html, 'html.parser')

            # Gömülü JSON verisini çıkarıyoruz
            script_tag = soup.find('script', {'id': '__NEXT_DATA__', 'type': 'application/json'})
            if script_tag:
                json_data = json.loads(script_tag.string)  # JSON verisini ayrıştırıyoruz

                # Kitap bilgilerini alıyoruz
                books = json_data['props']['pageProps']['response']['_sonuc']['gonderiler']

                current_books = set()  # Bu sayfadaki kitapların seti
                for book in books:
                    if 'adi' in book and 'ekBilgi' in book:  # Kitap adı ve ek bilgi varsa
                        book_name = book['adi']
                        if book_name in all_books_set:
                            continue  # Kitap zaten eklenmişse atla

                        ek_bilgi = book['ekBilgi']
                        # Kullanıcının verdiği puanı ek bilgi içinden çıkarıyoruz
                        user_rating_match = re.search(r'· (\d+)/10 puan verdi', ek_bilgi)
                        user_rating = user_rating_match.group(1) if user_rating_match else "Puan yok"

                        books_data.append((book_name, user_rating))
                        all_books_set.add(book_name)  # Kitabı tüm kitaplar setine ekle
                        current_books.add(book_name)  # Kitap adını bu sayfanın setine ekle

                # Eğer bu sayfadaki kitaplar bir önceki sayfadakilerle aynıysa döngüyü sonlandır
                if current_books == previous_books:
                    print("Kitaplar aynı, döngü sonlandırılıyor.")
                    break

                # Bir sonraki iterasyon için mevcut sayfanın kitaplarını sakla
                previous_books = current_books

            current_page += 1  # Sonraki sayfaya geç
        time.sleep(1)
        driver.quit()

        # Sonuçları yazdır
        for book_name, user_rating in books_data:
            print(f"Kitap Adı: {book_name}, Kullanıcı Puanı: {user_rating}")
            
        with open(csv_file_name, mode="a", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)

            # Kullanıcı adını URL'den çıkar
            user_name = base_url.split("/")[3]

            # Kitap verilerini yaz (başlıklar yazılmıyor, çünkü dosya zaten var)
            for book_name, user_rating in books_data:
                writer.writerow([user_name, book_name, user_rating])

        print(f"Yeni veriler '{csv_file_name}' dosyasına eklendi.")
    else:
        print(f"Kullanıcı atlanıyor: {base_url}")
        driver.quit()
        continue  # Bu kullanıcıyı atla


```
